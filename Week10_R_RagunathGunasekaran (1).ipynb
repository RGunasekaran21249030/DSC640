```r
# ggplot2 examples
library(ggplot2) 
```

## Histogram Chart

```r
Boston_housing_df=read.csv("Boston_housing.csv")
irisdata_df=read.csv("irisdata.csv")
```

```r
colors = c("black", "red", "green", "yellow","pink", "brown","maroon","grey","blue","orange") 
hist(Boston_housing_df$age, main = "Age - Histogram", xlab = "Age", ylab = "Frequency", col = colors)
```

## Box Plot

```r
# Boxplots of mpg by number of gears 
# observations (points) are overlayed and jittered
qplot(gear, mpg, data=mtcars, geom=c("boxplot", "jitter"), 
   fill=gear, main="Mileage by Gear Number",
   xlab="", ylab="Miles per Gallon")
```

```r
ggplot(irisdata_df, aes(Species, PetalLengthCm, fill=Species)) + 
  geom_boxplot()+
  scale_y_continuous("Petal Length", breaks= seq(0,30, by=.5))+
  labs(title = "Iris - Box Plot", x = "Species")
```

## Scatter plot with point and Smooth

```r
ggplot(data = irisdata_df, aes(x = PetalLengthCm, y = PetalWidthCm))+
  xlab("Petal Length")+
  ylab("Petal Width") +
  geom_point(aes(color = Species,shape=Species))+
  geom_smooth(method='lm')+
  ggtitle("Petal Length vs Width")
```
